```python
from collections import defaultdict

def result(inputfile):
    lines = []
    neighbours = defaultdict(lambda: defaultdict(int))
    with open(inputfile) as f:
        for line in f:
            line = line.strip()
            lines.append(line)
    for i in range(len(lines)):
        for j in range(len(line)):
            if lines[i][j] == '@':
                for di in [-1, 0, 1]:
                    for dj in [-1, 0, 1]:
                        if di == 0 and dj == 0:
                            continue
                        ni = i + di
                        nj = j + dj
                        if 0 <= ni < len(lines) and 0 <= nj < len(line):
                            neighbours[ni][nj] += 1
    result = 0
    for i in range(len(lines)):
        for j in range(len(line)):
            if lines[i][j] == '@' and neighbours[i][j] < 4:
                result += 1
    return result
```
