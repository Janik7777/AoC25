```python
def result(inputfile):
    ranges_lower = []
    ranges_upper = []
    result = 0
    readingRanges = True
    with open(inputfile) as f:
        for line in f:
            if readingRanges:
                if line.strip() == "":
                    readingRanges = False
                    continue
                parts = line.strip().split("-")
                ranges_lower.append(int(parts[0]))
                ranges_upper.append(int(parts[1]))
            else:
                ingredient = int(line.strip())
                for i in range(len(ranges_lower)):
                    if ranges_lower[i] <= ingredient <= ranges_upper[i]:
                        result += 1
                        break
    return result
```

```python
def result2(inputfile):
    ranges_merged = []

    def merge_ranges(new_range):
        for i, range in enumerate(ranges_merged):
            if new_range[1] < range[0] - 1:
                continue
            if new_range[0] > range[1] + 1:
                continue
            new_range[0] = min(new_range[0], range[0])
            new_range[1] = max(new_range[1], range[1])
            del ranges_merged[i]
            return merge_ranges(new_range)
        ranges_merged.append(new_range)

    with open(inputfile) as f:
        for line in f:
            if line.strip() == "":
                return sum(r[1] - r[0] + 1 for r in ranges_merged)
            parts = line.strip().split("-")
            current_range = [int(parts[0]), int(parts[1])]
            merge_ranges(current_range)
```

```python
result2("example-input.txt")
```
