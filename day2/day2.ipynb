```python
def result(inputfile):
    result = 0
    with open(inputfile) as f:
        ranges = f.readline().strip().split(',')
        for rangeString in ranges:
            range_parts = rangeString.split('-')
            start = int(range_parts[0])
            end = int(range_parts[1])
            for value in range(start, end + 1):
                digits = len(str(value))
                if digits % 2 == 1:
                    continue
                if str(value)[:digits // 2] == str(value)[digits // 2:]:
                    result += value        
    return result
```

```python
def result2(inputfile):
    result = 0
    with open(inputfile) as f:
        ranges = f.readline().strip().split(',')
        for rangeString in ranges:
            range_parts = rangeString.split('-')
            start = int(range_parts[0])
            end = int(range_parts[1])
            for value in range(start, end + 1):
                result += value_check(value)

    return result

def value_check(value):
    digits = len(str(value))
    for i in range(1, digits // 2 + 1):
        if digits % i != 0:
            continue
        if i_check(value, i, digits):
            return value
    return 0

def i_check(value, i, digits):
    pattern = str(value)[:i]
    for j in range(i, digits, i):
        if str(value)[j:j+i] != pattern:
            return False
    return True
```
