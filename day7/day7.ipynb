```python

def result(inputfile):
    lines = []
    with open(inputfile) as f:
        for line in f:
            if line == '':
                break
            lines.append([*line.strip()])

    result = 0

    for i in range(1, len(lines)):
        for j in range(len(lines[i])):
            if lines[i - 1][j] == 'S' or lines[i - 1][j] == '|':
                if lines[i][j] == '.':
                    lines[i][j] = '|'
                elif lines[i][j] == '^':
                    lines[i][j - 1] = '|'
                    lines[i][j + 1] = '|'
                    result += 1
    # for i in range(len(lines)):
    #     print("".join(lines[i]))
    return result
```

```python
result("example-input.txt")
```

```python
def result2(inputfile):
    lines = []
    with open(inputfile) as f:
        for line in f:
            if line == '':
                break
            lines.append([*line.strip()])

    for i in range(len(lines)):
        for j in range(len(lines[i])):
            if lines[i][j] == 'S':
                    lines[i][j] = 1
            if lines[i][j] == '.':
                    lines[i][j] = 0
    
    for i in range(1, len(lines)):
        for j in range(len(lines[i])):
            if lines[i - 1][j] != '^' and lines[i - 1][j] != '0':
                if lines[i][j] == '^':
                    lines[i][j - 1] += lines[i - 1][j]
                    lines[i][j + 1] += lines[i - 1][j]
                else:
                    lines[i][j] += lines[i - 1][j]
    # for i in range(len(lines)):
    #     print(lines[i])
    return sum(lines[-1])
```

```python
result2("example-input.txt")
```
