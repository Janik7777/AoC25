```python
def result(inputfile):
    result = 0
    with open(inputfile) as f:
        for line in f:
            line = line.strip()
            current = 0
            for i in range(len(line)):
                digit1 = int(line[i])
                for j in range(i + 1, len(line)):
                    digit2 = int(line[j])
                    joltage = digit1 * 10 + digit2
                    if joltage > current:
                        current = joltage
            result += current
    return result
```

```python
def result2(inputfile):
    result = 0
    with open(inputfile) as f:
        for line in f:
            line = line.strip()
            lineJoltage = ''
            lastChar = 0
            for j in range(12, 0, -1):
                current = 0
                for i in range(lastChar, len(line) - j + 1):
                    digit1 = int(line[i])
                    if digit1 > current:
                        current = digit1
                        lastChar = i + 1
                lineJoltage += str(current)
            result += int(lineJoltage)
    return result
```

```python
result2("example-input.txt")
```
