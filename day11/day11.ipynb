```python
import networkx as nx
from collections import defaultdict

def result(inputfile):
    nodes = []
    edges = []
    with open(inputfile) as f:
        for line in f:
            if line == '':
                break
            parts = line.split(': ')
            nodes.append(parts[0])
            for edge in parts[1].strip().split(' '):
                edges.append( [parts[0], edge] )

    G = nx.DiGraph()
    G.add_nodes_from(nodes)
    G.add_edges_from(edges)
    
    ordering = list(nx.topological_sort(G))
    counts = defaultdict(int)
    counts['you'] = 1
    for node in ordering:
        for next in G.neighbors(node):
            counts[next] += counts[node]
    return counts['out']
```

```python
def result2(inputfile):
    nodes = []
    edges = []
    with open(inputfile) as f:
        for line in f:
            if line == '':
                break
            parts = line.split(': ')
            nodes.append(parts[0])
            for edge in parts[1].strip().split(' '):
                edges.append( [parts[0], edge] )

    G = nx.DiGraph()
    G.add_nodes_from(nodes)
    G.add_edges_from(edges)
    
    ordering = list(nx.topological_sort(G))
    
    svrToFftCounts = defaultdict(int)
    svrToDacCounts = defaultdict(int)
    fftToDacCounts = defaultdict(int)
    dacToFftCounts = defaultdict(int)
    fftToOutCounts = defaultdict(int)
    dacToOutCounts = defaultdict(int)
    svrToFftCounts['svr'] = 1
    svrToDacCounts['svr'] = 1
    fftToDacCounts['fft'] = 1
    dacToFftCounts['dac'] = 1
    fftToOutCounts['fft'] = 1
    dacToOutCounts['dac'] = 1
    for node in ordering:
        for next in G.neighbors(node):
            if next != 'dac':
                svrToFftCounts[next] += svrToFftCounts[node]
                fftToOutCounts[next] += fftToOutCounts[node]
            if next != 'fft':
                svrToDacCounts[next] += svrToDacCounts[node]
                dacToOutCounts[next] += dacToOutCounts[node]
            dacToFftCounts[next] += dacToFftCounts[node]
            fftToDacCounts[next] += fftToDacCounts[node]

    amoutSvrFftDacOut = svrToFftCounts['fft'] * fftToDacCounts['dac'] * dacToOutCounts['out']
    amoutSvrDacFftOut = svrToDacCounts['dac'] * dacToFftCounts['fft'] * fftToOutCounts['out']
    return amoutSvrFftDacOut + amoutSvrDacFftOut
    
```
