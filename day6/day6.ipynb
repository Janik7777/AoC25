```python
from collections import defaultdict
import math

def result(inputfile):
    numbers = defaultdict(lambda: defaultdict(int))
    with open(inputfile) as f:
        j = 0
        for line in f:
            if line.startswith('*'):
                continue
            numberStrings = line.split(' ')
            i = 0
            for numberString in numberStrings:
                if numberString.strip() != '':
                    numbers[i][j] = int(numberString.strip())
                    i += 1
            j += 1
    operators = line.strip().split(' ')
    k = 0
    result = 0
    length = len(numbers[0])
    for operator in operators:
        operator = operator.strip()
        if operator == '':
            continue
        #print([numbers[k][j] for j in range(length)])
        if operator == '+':
            result += sum(numbers[k][j] for j in range(length))

        elif operator == '*':
            result += math.prod(numbers[k][j] for j in range(length))
        k += 1
    return result
```

```python
def result2(inputfile):
    numbers = []
    with open(inputfile) as f:
        for line in f:
            if line.startswith('*'):
                continue
            numbers.append([*line])
    
            
    operators = [*line]
    result = 0
    numberLen = len(numbers)
    current = 0
    currentOp = ""
    for i in range(len(operators)):
        number= ''.join([numbers[j][i] for j in range(numberLen)]).strip()
        if operators[i] == '+':
            current = int(number)
            currentOp = '+'
        elif operators[i] == '*':
            current = int(number)
            currentOp = '*'
        else:
            if number == '':
                result += current
                current = 0
            elif currentOp == '+':
                current += int(number)
            elif currentOp == '*':
                current *= int(number)
    return result + current
```
