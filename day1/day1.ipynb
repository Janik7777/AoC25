```python
def result(inputfile):
    value = 50
    result = 0
    with open(inputfile) as f:
        for line in f:
            line = line.strip()
            signum = 1 if line[0] == 'R' else -1
            distance = int(line[1:])
            value = (value + signum * distance) % 100
            if value == 0:
                result += 1
        
    return result
```

```python
def result2(inputfile):
    value = 50
    result = 0
    with open(inputfile) as f:
        for line in f:
            line = line.strip()
            signum = 1 if line[0] == 'R' else -1
            distance = int(line[1:])
            newvalue = (value + signum * distance)

            if newvalue >= 100:
                result += newvalue // 100
                value = newvalue % 100
            elif newvalue <= 0:
                result += (-newvalue) // 100 + 1
                if value == 0:
                    result -= 1
                value = newvalue % 100
            else:
                value = newvalue % 100
    return result
```

```python
result2("example-input.txt")
```
